```python
# [특정 키워드를 입력했을 때 네이버에서 내 아이디가 어디에 위치해 있는가?]


# ### 특정 키워드와 아이디 입력


# ### 지식스니펫 또는 스마트블럭에 내 아이디가 있는가?

# 아이디가 있으면 아이디 반환
# 없으면 '블로그' 탭으로 이동


# ### 블로그 탭에서 아이디가 어떤 페이지에 있는지 확인

# =

# 1. 특정 키워드와 아이디 입력 시, 해당 아이디가 스마트블럭에 있는지 확인


# 키워드 입력

# url에 키워드 입력

# 스마트블럭 공간에 아이디가 있는지 확인

# 2. 특정 키워드와 아이디 입력 시, 해당 아이디가 지식스니펫에 있는지 확인
```

최종 코드

```python
### 키워드 입력

from selenium import webdriver

driver = webdriver.Chrome()

keyword = input("키워드를 입력해 주세요")

driver.get(f"https://search.naver.com/search.naver?ssc=tab.nx.all&where=nexearch&sm=tab_jum&query={keyword}")


### 아이디 저장

from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

dongju_id_list = [
    "designersiun", "singsong0514", "phoenixjeong", "hamas3000", "roses777",
    "dongjulaw1", "dongjulaw2", "dongjusuwon1", "dongjulaw6", "dj_ehdwn1",
    "rudnfdldi00", "ehtlarhdwn", "widance", "yellowoi", "dongjulaw",
    "tale1396", "dongjulaw5", "dongjulaw100", "dongjulaw4", "dongjulaw02",
    "dksro018", "cckjjt", "qusghtkehdwn", "dongjulaw7", "ujm159",
    "dong-ju-law", "dongjulaw3", "ehdwnfh", "kkobugi39"
]


### 지식스니펫 확인
from selenium.webdriver.common.by import By

knowledge_snippet = driver.find_element(By.CSS_SELECTOR, '.source_box .txt.elss').get_attribute('href')
split_knowledge_snippet = knowledge_snippet.split('/')[-1]

if split_knowledge_snippet in dongju_id_list:
    print(f"스니펫입니다 : {split_knowledge_snippet}")
else:
    smartblock_research= soup.select_one('.BZppu7wV32H2scXPRUVx.fds-info-inner-text').attrs['href']


### 스마트블럭 확인
smartblock_research= soup.select_one('.BZppu7wV32H2scXPRUVx.fds-info-inner-text').attrs['href']

# 하나씩 넣어서 검사해 본다.
# 검사해서 있으면, 해당 값을 특정 변수에 저장한다.
# 그리고 반복을 중단한다.
found_id = None
for id in dongju_id_list:
    if id in smartblock_research:
        found_id = id
        break

# 그런 다음 특정변수에 값이 들어가서 이를 충족한다는 조건문을 작성한다.
# 참이면 아이디를 반환
# 거짓이면 스블이 아닙니다 반환.
if found_id:
    print(f"스블입니다 : {found_id}")
else:
    ### 블로그 탭 클릭
    driver.find_element(By.CSS_SELECTOR, '.flick_bx:nth-of-type(3) > a').click()


### 무한스크롤 처리
import time
import random

random_time = random.uniform(1, 1.5)

while True :
    before_height = driver.execute_script("return document.body.scrollHeight")

    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(random_time)

    after_height = driver.execute_script("return document.body.scrollHeight")

    if before_height == after_height:
        break

    before_height = after_height

### 블로그 순위 체크
blog_ids = driver.find_elements(By.CSS_SELECTOR, '.user_info a')
for blog_id in blog_ids:
    href = blog_id.get_attribute('href')
    extracted_id = href.split('/')[-1]

    if extracted_id in dongju_id_list:
        index = dongju_id_list.index(extracted_id) + 1
        print(f"{extracted_id} : {index}위")
    else:
        print(f"{extracted_id} : 순위밖")
```
